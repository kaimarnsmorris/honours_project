# Hamiltonian Monte Carlo

Let us first attempt to implement the Leapfrog integrator and the Hamiltonian dynamics.

```python
import numpy as np
import scipy.stats as stats
```

```python
def leapfrog(q, p, grad_U, step_size, num_steps):
    """Perform the leapfrog integration."""
    p -= 0.5 * step_size * grad_U(q)
    for _ in range(num_steps):
        q += step_size * p
        p -= step_size * grad_U(q)
    return q, p
```
